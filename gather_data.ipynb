```python
import numpy as np
import pandas as pd
from spotipy.oauth2 import SpotifyClientCredentials
import spotipy
import time
from tqdm import tqdm
import json
```

```python
client_id = ""
client_secret = ""

client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
```

```python
# Audio Features
dance = []
energy = []
key = []
loudness = []
speech = []
acoustic = []
instrument = []
live = []
valence = []
tempo = []
duration = []
time = []

#Track Features
t_name = []
album_name = []
t_pop = []

#Artist Features
a_name = []
a_pop = []
a_genre = []
a_follower = []

#id
u_id = []

#Release date
date = []

permalink = ["https://open.spotify.com/playlist/37i9dQZEVXbNG2KDcFcKOF?si=1333723a6eff4b7f",
            "https://open.spotify.com/playlist/6DsHDZmrOgLBmxa9DJE6sb?si=305a750fc30f4af5",
            "https://open.spotify.com/playlist/37i9dQZEVXbKXQ4mDTEBXq?si=de15894c2012456f",
            "https://open.spotify.com/playlist/2IVy9JCLWgDE7a0fqocFrO?si=ff552f76c60844d4",
            "https://open.spotify.com/playlist/6Hj8E9Vdadx8hsBhEpXYmy?si=b90b15c8ab1b4f80",
            "https://open.spotify.com/playlist/37i9dQZF1DWZd79rJ6a7lp?si=6b57183da0f54098",
            "https://open.spotify.com/playlist/5hAhTBlPO5wjlOzTzpSyYs?si=0623ea8011714e4c",
            "https://open.spotify.com/playlist/0YXSaCQf45KtpAtMIkH3us?si=22f78eb0215542eb",
            "https://open.spotify.com/playlist/2WWIP4h23cnKtRv3QSA6WG?si=56f4b76fb42a4998",
            "https://open.spotify.com/playlist/5XHOt7ufAj0lPjXcarzSJe?si=3b18e52313954337",
            "https://open.spotify.com/playlist/4kfZqH5GFkKoMI82WDIMjf?si=95d007ed8eb24613",
            "https://open.spotify.com/playlist/25VfAePi0FitCJEckcwzDE?si=324f0507fd2f43c1",
            "https://open.spotify.com/playlist/37i9dQZF1DWYoioDxgMsjZ?si=aa8eb07da1a54335",
            # "https://open.spotify.com/playlist/4mIvDWKkWgwPwp1zMdjelo?si=bf80cfbe791c4248",
            # "https://open.spotify.com/playlist/6j6e8b297sIj82dEWTAJ2Y?si=4d00b03989144c52",
            # "https://open.spotify.com/playlist/65cSYpaqgkfg2PvyZC69fp?si=faaad66a2ef14ac0",
            # "https://open.spotify.com/playlist/2bCSK34MFjSj66xi1eeQrY?si=86c9c4259f874fd2",
            # "https://open.spotify.com/playlist/4sL2mzgFBBe7gzmK5Emzrb?si=b833700751ea42a7",
            # "https://open.spotify.com/playlist/37i9dQZF1DX0XUsuxWHRQd?si=e5776e74471f4a92",
            # "https://open.spotify.com/playlist/4idNGo1kkfMFeonTeupH9p?si=c5137ce2eacd4550",
            # "https://open.spotify.com/playlist/37i9dQZF1DX89EkrAT8Z6U?si=04a0e53538664eee",
            # "https://open.spotify.com/playlist/37i9dQZF1DX5Vy6DFOcx00?si=8387aa2a7096481e",
            # "https://open.spotify.com/playlist/37i9dQZEVXbLiRSasKsNU9?si=818f320cc030464a",
            # "https://open.spotify.com/playlist/37i9dQZEVXbNZbJ6TZelCq?si=7956265886c443b9",
            # "https://open.spotify.com/playlist/49Y8YgoKoPFEELOeEiQrLx?si=9e9e8a28e221457a",
            ]

def get_playlist_tracks(playlist_id):
    results = sp.playlist_items(playlist_id)
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks

for link in permalink:
    link = link.split("/")[-1].split("?")[0]
    playlist_tracks = get_playlist_tracks(link)

    # print(json.dumps(playlist_tracks, indent=2))
    tracks_playlist = sp.playlist_items(link)
    print("Tracklist pt :", link)
    for track in tqdm(tracks_playlist["items"]):
        try:
            track_uri = track["track"]["uri"]
            features = sp.audio_features(track_uri)[0]
            d = features["danceability"]
            e = features["energy"]
            k = features["key"]
            l = features["loudness"]
            s = features["speechiness"]
            a = features["acousticness"]
            i = features["instrumentalness"]
            li = features["liveness"]
            v = features["valence"]
            t = features["tempo"]
            d = features["duration_ms"]
            tm = features["time_signature"]

            dance.append(d)
            energy.append(e)
            key.append(k)
            loudness.append(l)
            speech.append(s)
            acoustic.append(a)
            instrument.append(i)
            live.append(li)
            valence.append(v)
            tempo.append(t)
            duration.append(d)
            time.append(tm)

            # Release Date
            release_date = track["track"]["album"]["release_date"]
            date.append(release_date)
            
            #ID
            user_id = track["track"]["id"]
            u_id.append(user_id)
            
            #Track name
            track_name = track["track"]["name"]
            t_name.append(track_name)

            #Album
            album = track["track"]["album"]["name"]
            album_name.append(album)

            #Popularity of the track
            track_pop = track["track"]["popularity"]
            t_pop.append(track_pop)

            #Main Artist
            artist_uri = track["track"]["artists"][0]["uri"]
            artist_info = sp.artist(artist_uri)


            #Name, popularity, genre, followers
            artist_name = track["track"]["artists"][0]["name"]
            a_name.append(artist_name)

            artist_pop = artist_info["popularity"]
            a_pop.append(artist_pop)

            artist_genres = artist_info["genres"]
            a_genre.append(artist_genres)

            artist_followers = artist_info["followers"]["total"]
            a_follower.append(artist_followers)
            
        except:
            pass
    
    while tracks_playlist['next']:
        results = sp.next(tracks_playlist)
        for track in tqdm(results["items"]):
            try:
                track_uri = track["track"]["uri"]
                features = sp.audio_features(track_uri)[0]
                d = features["danceability"]
                e = features["energy"]
                k = features["key"]
                l = features["loudness"]
                s = features["speechiness"]
                a = features["acousticness"]
                i = features["instrumentalness"]
                li = features["liveness"]
                v = features["valence"]
                t = features["tempo"]
                d = features["duration_ms"]
                tm = features["time_signature"]

                dance.append(d)
                energy.append(e)
                key.append(k)
                loudness.append(l)
                speech.append(s)
                acoustic.append(a)
                instrument.append(i)
                live.append(li)
                valence.append(v)
                tempo.append(t)
                duration.append(d)
                time.append(tm)

                # Release Date
                release_date = track["track"]["album"]["release_date"]
                date.append(release_date)
                
                #ID
                user_id = track["track"]["id"]
                u_id.append(user_id)
                
                #Track name
                track_name = track["track"]["name"]
                t_name.append(track_name)

                #Album
                album = track["track"]["album"]["name"]
                album_name.append(album)

                #Popularity of the track
                track_pop = track["track"]["popularity"]
                t_pop.append(track_pop)

                #Main Artist
                artist_uri = track["track"]["artists"][0]["uri"]
                artist_info = sp.artist(artist_uri)


                #Name, popularity, genre, followers
                artist_name = track["track"]["artists"][0]["name"]
                a_name.append(artist_name)

                artist_pop = artist_info["popularity"]
                a_pop.append(artist_pop)

                artist_genres = artist_info["genres"]
                a_genre.append(artist_genres)

                artist_followers = artist_info["followers"]["total"]
                a_follower.append(artist_followers)
                
            except:
                pass


    spotify_df = pd.DataFrame({'id':u_id, 'artist_name' : a_name, 'track_name' : t_name, 'album_name' : album_name,
                             'artist_genre' : a_genre, 'release_date' : date, 'artist_popularity' : a_pop, 'track_popularity' : t_pop,
                             'artist_followers' : a_follower, 'danceability' : dance, 'energy' : energy,
                            'key' : key, 'loudness' : loudness, 'speechiness' : speech, 'acousticness' : acoustic,
                            'instrumentalness' : instrument, 'liveness' : live, 'valence' : valence, 'tempo' : tempo,
                            'duration_ms' : duration, 'time_signature' : time})
print(spotify_df.shape)
spotify_df.head()
```
